```python
import matplotlib.pyplot as plt
import math
import numpy as np
import matplotlib as mpl
from matplotlib import rc
import seaborn as sns
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed

%matplotlib inline
```

```python

def ht(x):
    if abs(x) <= alphat:
        return C1 * alphat**2 / 4
    elif abs(x) >= alphat and abs(x) <= 1.5 * alphat:
        return -C1 * (abs(x) - alphat)**2 / 2 + C1 * alphat**2 / 4
    elif abs(x) >= 1.5 * alphat and abs(x) <= 2 * alphat:
        return C1 * (abs(x) - 2 * alphat)**2 / 2
    else:
        return 0
    
```

```python
C1 = 1.0
alphat = 0.25

X = [0.01 * (i - 100) for i in range(0, 201)]

Y = [ht(x) for x in X]

fig, ax = plt.subplots(figsize=(6,4))

plt.plot(X, Y, lw=4.0)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=20)
ax.yaxis.grid(True, lw=0.4)
plt.xlabel(r'$x$',  fontsize=28)
plt.ylabel(r'$h_t(x)$',  fontsize=28)
ax.set_title(r'$C_1 = 1.0, \alpha_t = 0.25$', fontsize=28)

plt.tight_layout()
plt.savefig('bad_ex_1.pdf', format='pdf', dpi=100)
```
