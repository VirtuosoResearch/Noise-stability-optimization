```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
import matplotlib as mpl
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
directory = './logs/'

def get_results(filename):
    with open(directory + filename, 'r') as f:
        lines = f.readlines()
        
        results = {'epochs': [], 'train_loss': [], 'val_loss': [], 'dist': []}
        for line in lines[1:]:
            res = line.split(' ')

            if int(res[1]) > 5000:
                continue
            results['epochs'].append(int(res[1]))
            results['train_loss'].append(float(res[3]))
            results['val_loss'].append(float(res[5]))
            results['dist'].append(float(res[7]) ** 2)
            
    return results
```

```python
gd = get_results('gd')

nso = get_results('nso008')

wpsgd = get_results('wpsgd008')
```

```python
epochs = [i * 20 for i in range(0, )]

plt.figure(figsize=(6,5))


plt.plot(gd['epochs'], gd['train_loss'], c='k', ls='--', lw=4, label=r'$\mathrm{GD}$')
plt.plot(nso['epochs'], nso['train_loss'], c='r', lw=4, label=r'$\mathrm{NSO}$' + r'~$(k = 1)$')
#plt.plot(gd['epochs'], gd['val_loss'], c='m', lw=2, label='GD, val loss')
#plt.plot(nso['epochs'], nso['val_loss'], c='b', lw=2, label='NSO, val loss')
plt.plot(wpsgd['epochs'], wpsgd['train_loss'], c='b', ls='-.', lw=4, label=r'$\mathrm{WP}$'+'-'+r'$\mathrm{GD}$')
#plt.plot(wpsgd['epochs'], wpsgd['val_loss'], c='m', ls='--', lw=2, label='WPSGD, val loss')
plt.xlabel(r'$\mathrm{Number~of~epochs}$', fontsize=24)
plt.ylabel(r'$\mathrm{Training~MSE}$', fontsize=24)
#plt.yscale('log')
plt.xticks([0, 1000, 2000, 3000, 4000, 5000], fontsize=20)
plt.yticks(fontsize=20)
plt.grid(ls = ':', lw=0.8)
plt.legend(fontsize=24)#,loc=3)
plt.tight_layout()
plt.savefig('compare_algs_ms_train.eps')
```

```python
epochs = [i * 20 for i in range(0, )]

plt.figure(figsize=(6,5))


plt.plot(gd['epochs'], gd['dist'], c='k', ls='--', lw=4, label=r'$\mathrm{GD}$')
plt.plot(nso['epochs'], nso['dist'], c='r', lw=4, label=r'$\mathrm{NSO}$')
#plt.plot(gd['epochs'], gd['val_loss'], c='m', lw=2, label='GD, val loss')
#plt.plot(nso['epochs'], nso['val_loss'], c='b', lw=2, label='NSO, val loss')
plt.plot(wpsgd['epochs'], wpsgd['dist'], c='b', ls='-.', lw=4, label=r'$\mathrm{WP}$'+'-'+r'$\mathrm{GD}$')
#plt.plot(wpsgd['epochs'], wpsgd['val_loss'], c='m', ls='--', lw=2, label='WPSGD, val loss')
plt.xlabel(r'$\mathrm{Number~of~epochs}$', fontsize=24)
plt.ylabel(r'$\mathrm{Euclidean~Distance}$', fontsize=24)
plt.yscale('log')
plt.xticks([0, 1000, 2000, 3000, 4000, 5000], fontsize=20)
plt.yticks(fontsize=20)
plt.grid(ls = ':', lw=0.8)
plt.legend(fontsize=24)#,loc=3)
plt.tight_layout()
plt.savefig('compare_algs_ms_dist.eps')
```
